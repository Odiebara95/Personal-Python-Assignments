```python
import numpy as np
```

```python
with open("../DS-Training-2022/datasets/alcohol/estimated_numbers.csv","r") as file:
    content= file.readlines()
    print(content)
```

```python
for case in content[1:]:
    print(round(float(case.split(",")[4])))
```

```python
cases_median=[]
countries=[]
for case in content[1:]:
    try:
        countries.append(case.split(",")[0])
        cases_median.append(round(float(case.split(",")[4])))
        
    except ValueError:
        cases_median.append(0)
cases_median= np.array(cases_median)
countries= np.array(countries)
print(cases_median)
print(countries)
```

```python
len(cases_median)
```

```python
len(countries)
```

```python
from matplotlib import pyplot as plt
```

```python
len(np.unique(countries))
```

```python
plt.figure(figsize=(20,20))
plt.barh(countries,cases_median)
plt.grid(3)
plt.title("Graph of expected cases of malaria by country")
plt.xlabel("cases_median")
plt.ylabel("Country")
```

```python
number_of_deaths=[]
countries=[]
for case in content[1:]:
        countries.append(case.split(",")[0])
        number_of_deaths.append(case.split(",")[2])
number_of_deaths= np.array(number_of_deaths)
countries= np.array(countries)
print(number_of_deaths)
print(countries)
```

```python
deaths=[]
countries=[]
for case in content[1:]:
        countries.append(case.split(",")[0])
        deaths.append(case.split(",")[2])
number_of_deaths = [i.split('[',1)[0] for i in deaths]
number_of_deaths= np.array(number_of_deaths)
countries= np.array(countries)
print(number_of_deaths)
print(countries)
```

```python
len(number_of_deaths)
```

```python
plt.figure(figsize=(20,20))
plt.barh(countries,number_of_deaths)
plt.grid(3)
plt.title("Graph of reported cases of malaria by country")
plt.xlabel("number_of_deaths")
plt.ylabel("Country")
```
