```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# graph for ùë¶=ùëêùëúùë†(ùë•)
x=np.linspace(-10,10,1000)
cosine=np.cos(x) 
plt.plot(x,cosine)
plt.grid(20)
```

```python
# graph for ùë¶=tan(ùë•)

x=np.linspace(-10,10,1000)
tangent=np.tan(x) 
plt.plot(x,tangent)
plt.grid(20)
```

```python
# graph for ùë¶=2ùë†ùëñùëõ(4ùë•)‚àíùëêùëúùë†(3ùë•)
x=np.linspace(-10,10,1000)
y=np.sin(4*x)*2 - np.cos(3*x)
plt.plot(x,y)
plt.grid(20)
```

```python
# graph for y= e**x / (1+ e**x)
x=np.linspace(-10,10,1000)
y=np.exp(x) / (1+ np.exp(x))
plt.plot(x,y)
plt.grid(20)
```
